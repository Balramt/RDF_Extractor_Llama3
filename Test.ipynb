```python
# Configuration
LOCAL_MODEL = "llama3:latest"
RDF_FILEPATH = r"D:\New folder\RDF_Extractor_Llama3\Barack-Obama.txt"  # Update this with your Text file path
```

```python
# Update paths for input and output files
JSONL_FILEPATH = r"D:\New folder\RDF_Extractor_Llama3\Data\ont_1_movie_prompts.JSONL"
OUTPUT_FILEPATH = r"D:\New folder\RDF_Extractor_Llama3\Data\Llama3_Output\generated_output.jsonl"
```

```python
# Initialize the language model
llm = Ollama(base_url="http://localhost:11434", model=LOCAL_MODEL, verbose=True)
```

```python
# Load the JSONL file containing prompts
with jsonlines.open(JSONL_FILEPATH) as reader:
    prompts_data = list(reader)
```

```python
# List to store the output results
results = []
# Process each prompt
for item in prompts_data:
    prompt_id = item['id']
    prompt_text = item['prompt']
    #print(prompt_id)
    # Generate output using the Llama model
    response = llm.generate([prompt_text], max_length=900)
    # Extract generated text
    if response and response.generations:
        generated_text = response.generations[0][0].text
    else:
        generated_text = "No text generated."
    # Store the result
    result = {
        "id": prompt_id,
        "prompt": prompt_text,
        "generated_text": generated_text
    }
    results.append(result)

    # Print the generated summary (optional)
    print(f"ID: {prompt_id}\nGenerated Summary:\n{generated_text}\n{'-'*50}")
```

```python
# Save the results to a new JSONL file
with jsonlines.open(OUTPUT_FILEPATH, mode='w') as writer:
    writer.write_all(results)

print(f"Results saved to {OUTPUT_FILEPATH}")
```

```python
def process_llama_output(output):
    # Split the output by ID segments
    segments = output.strip().split('\n--------------------------------------------------\n')
    
    results = []
    for segment in segments:
        # Extract ID and summary
        id_match = re.search(r'ID: (\S+)', segment)
        if id_match:
            id_value = id_match.group(1)
        else:
            continue
        
        # Extract the summary part of the segment
        summary_match = re.search(r'Generated Summary:\n([\s\S]+)', segment)
        if summary_match:
            summary_text = summary_match.group(1)
            
            # Extract the triples
            triples = re.findall(r'\* (\w+)\(([^,]+),([^)]*)\)', summary_text)
            
            # Format the response
            formatted_response = {
                "id": id_value,
                "response": '\n'.join(f"{triple[0]}({triple[1]},{triple[2]})" for triple in triples),
                "triples": [[triple[1], triple[0], triple[2]] for triple in triples]
            }
            
            results.append(formatted_response)
    
    return results
```

```python
# Function to read the JSONL file and extract ground truth data
def load_ground_truth(file_path, max_id=20):
    ground_truth = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            data = json.loads(line.strip())
            # Extract the numeric part of the ID and check if it's within the range
            numeric_id = int(data["id"].split("_")[-1])
            if numeric_id <= max_id:
                ground_truth.append(data)
    return ground_truth
```

```python
def normalize_string(s):
    return s.replace(" :", ":").replace(": ", ":").strip()
```

```python
# Function to calculate precision
def calculate_precision(llm_output, ground_truth):
    total_correct_matches = 0
    total_predicted = 0
    matching_triples = []

    # Convert ground truth triples to a dictionary for easier comparison
    ground_truth_dict = {}
    for entry in ground_truth:
        ground_truth_dict[entry["id"]] = [(normalize_string(t["sub"]), t["rel"].strip(), t["obj"].strip()) for t in entry["triples"]]

    for entry in llm_output:
        predicted_triples = [(normalize_string(t[0]), t[1].strip(), t[2].strip()) for t in entry["triples"]]
        total_predicted += len(predicted_triples)

        ground_truth_triples = ground_truth_dict.get(entry["id"], [])

        for triple in predicted_triples:
            if triple in ground_truth_triples:
                total_correct_matches += 1
                matching_triples.append(triple)

    # Calculate precision
    print("Total Correct Matches:", total_correct_matches)
    print("Total Predicted Triples:", total_predicted)
    print("Matching Triples:", matching_triples)
    
    precision = total_correct_matches / total_predicted if total_predicted > 0 else 0
    return precision
```
